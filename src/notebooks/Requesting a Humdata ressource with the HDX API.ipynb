# Requesting a Humdata ressource with the HDX API

From the [documentation](https://data.humdata.org/documentation). How to access data and metadata of a Humdata ressource.

```python
from hdx.hdx_configuration import Configuration 
from hdx.data.dataset import Dataset
from pathlib import Path
LowercasePath = lambda x: os.path.normcase(Path(x).absolute())
import os
```

```python
dataset_name = 'unicef-mnch-diarcare'
ressource_index = 1
```

```python
Configuration.create(hdx_site='prod', user_agent='D4G_ACF', hdx_read_only=True)
dataset = Dataset.read_from_hdx(dataset_name)
resources = dataset.get_resources()
```

```python
data_folder = LowercasePath("../../data/sources/")
url, path = resources[ressource_index].download(folder = data_folder)
print('Resource URL %s downloaded to %s' % (url, path))
```

```python
# Example reading
import pandas as pd
diarrhoea_df = pd.read_csv(path)
diarrhoea_df.head()
```
